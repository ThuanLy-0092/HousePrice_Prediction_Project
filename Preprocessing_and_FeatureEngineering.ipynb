# Sampling Methods

Khi tiến hành nghiên cứu về một nhóm đối tượng, ta hiếm khi có thể thu thập dữ liệu từ mọi đối tượng có trong nhóm đó. Thay vào đó, ta chọn một mẫu.  
Mẫu là nhóm các đối tượng sẽ thực sự tham gia vào quá trình nghiên cứu.
Để có thể rút ra kết luận có giá trị từ kết quả nghiên cứu, ta phải quyết định cẩn thận cách chọn mẫu đại diện cho toàn bộ nhóm. Đây được gọi là `phương pháp lấy mẫu (Sampling Methods)`  

Có hai loại phương pháp lấy mẫu chính mà ta có thể sử dụng trong nghiên cứu của mình:


1. **Lấy mẫu xác suất (Probability sampling):** bao gồm lựa chọn ngẫu nhiên, cho phép bạn đưa ra những suy luận thống kê chắc chắn về toàn bộ nhóm.  
>Lấy mẫu xác suất có nghĩa là mọi thành viên trong tổng thể đều có cơ hội được chọn. Nó chủ yếu được sử dụng trong nghiên cứu định lượng . Nếu bạn muốn tạo ra kết quả đại diện cho toàn bộ tổng thể thì kỹ thuật lấy mẫu xác suất là lựa chọn hợp lý nhất.



2. **Lấy mẫu phi xác suất (Non-probability sampling):**  bao gồm lựa chọn không ngẫu nhiên dựa trên sự thuận tiện hoặc các tiêu chí khác, cho phép bạn dễ dàng thu thập dữ liệu.  

   >Mẫu phi xác suất chọn các cá nhân dựa trên tiêu chí không ngẫu nhiên, khiến một số thành viên không có cơ hội được chọn. Loại mẫu này dễ tiếp cận và chi phí thấp hơn, nhưng `có nguy cơ sai lệch cao`, làm cho suy luận về tổng thể kém chính xác. Kỹ thuật này thường dùng trong nghiên cứu thăm dò và định tính để hiểu rõ một nhóm nhỏ hoặc ít được nghiên cứu, thay vì kiểm tra giả thuyết về một tổng thể rộng. Mục tiêu là làm mẫu đại diện cho tổng thể càng nhiều càng tốt.

## **NHẬN ĐỊNH**:

Công việc `Cào dữ liệu có chọn lọc` (**Selective Web Scraping**) **có thể được xem là một phương pháp `lấy mẫu phi xác suất`**. Trong cào dữ liệu có chọn lọc, các đối tượng được chọn dựa trên các tiêu chí cụ thể (*là các đặc trưng chính cần phải có đã được trình bày ở phần trên*) và không ngẫu nhiên, điều này phù hợp với định nghĩa của lấy mẫu phi xác suất. Mục tiêu là thu thập dữ liệu từ các nguồn cụ thể hoặc các nhóm đối tượng nhất định, thay vì chọn ngẫu nhiên từ tổng thể.

### Price Handling

Nhận thấy cột `Price` là giá của một bất động sản nhưng lại có kiểu dữ liệu là `object`, ta thực hiện chuyển đổi các dữ liệu ở cột này về giá trị số tương ứng

````python
data['Price'] = data['Price'].apply(lambda x: re.sub(r'(\d+\.\d+|\d+)\s*triệu', lambda y: str(float(y.group(1))/1000) + ' tỷ',x))
data['Price'] = data['Price'].apply(lambda x: re.sub('[^\d.]','',x))
````

````python
data.info()
````

Do có một số bản tin có giá bất động sản không hợp lệ hoặc sai hình thức.

VD: 3600000 tỷ,... thì ta sẽ xử lý nó về giá trị tương ứng hoặc xóa khỏi dataset.

````python
# Hàm chia giá trị cho 1 tỷ nếu nằm trong khoảng từ 9 đến 12 chữ số (Tức ở mức Trăm Triệu - Trăm Tỷ)
def process_price(price):
    try:
        price = float(price)
        if 9 <= len(str(price)) <= 12:
            return float(price / 1000000000)
        else:
            return price
    except:
        return None
````

Việc xử lý những giá trị là None sẽ được thực hiện ở quy trình ` Xử lý lại cột Price sau bước Geocoding`

````python
data['Price'] = data['Price'].apply(process_price)

# Loại bỏ các hàng có giá trị 'Price' là None
data = data.dropna(subset=['Price'])
````

### Geocoding

>Cột `Address`, `district` và `city/province` thể hiện địa chỉ của bất động sản, các dữ liệu ở các cột này ở dạng `object` (dạng chuỗi) để có thể đưa vào phân tích và trực quan ta chuyển về dạng `longitude` và `latitude` (hay `kinh độ` và `vĩ độ`), công việc này gọi là `mã hóa địa lý (Geocoding)`.

**Mã hóa địa lý (Geocoding) với API OpenStreetMap**

>**API OpenStreetMap miễn phí** nhưng sẽ chậm và kém chính xác hơn so với việc sử dụng API Google Maps, nhưng do điều kiện về tài nguyên của nhóm chúng tôi chỉ có khả năng sử dụng API OpenStreetMap.

>Với API OpenStreetMap thì việc xác định chính xác dựa trên địa chỉ `Address` sẽ gặp khó khăn vì **nó không có phạm vi địa chỉ tốt ở tất cả các khu vực**.

>Do đó chúng tôi sẽ sử dụng điều kiện nếu không thể xác định `kinh/vĩ độ (longitude/latitude)` từ `Address` thì sẽ sử dụng một các trường khác có ít thông tin hơn, đó là `district` và `city/province`. Với việc xác định địa chỉ với Quận/Huyện (district) của một Thành Phố/Tỉnh (city/province) thì sẽ dễ dàng và nhanh hơn nhưng đổi lại sẽ kém chính xác về địa chỉ của một bất động sản.

>Tuy nhiên, với việc sử dụng API OpenStreetMap 'miễn phí' chúng tôi luôn bị giới hạn truy cập và nhận được một số kết quả không mong muốn. Vì thế sẽ xuất hiện các giá trị `NULL` ở hai cột `longitude/latitude` mặc dù có đầy đủ `Address`.

**Vấn đề:**

>* API OpenStreetMap có giới hạn số lượng truy cập cho một tài khoản (<=5000): đối với một tập dữ liệu lớn hơn 8000 dòng thì việc dùng API một lần để Geocoding là không thể.

>* Thêm nữa sau mỗi lần Geocoding `Address` nếu không tìm ra `longitude/latitude` thì code còn phải thực hiện số lần `retries (thực hiện lại)` điều này cũng được tính thêm vào số lượt truy cập, qua đó nếu muốn Geocoding 1000 dòng thì số lượt truy cập API phải ít nhất 2000 lượt nếu `retries = 3` (Chưa tính đến số lượt Geocoding theo `district` và `city/province` nếu không Geocoding được theo `Address`.

>* Mỗi lượt Gecoding, chuỗi địa chỉ `Address` hoặc `district, city/province` sẽ được truyền vào để mã hóa, nhưng sẽ có những trường hợp chuỗi địa chỉ quá phức tạp không thể mã hóa để tìm ra longitude/latitude từ đó sẽ xuất hiện những lỗi. Khi lỗi xuất hiện quá nhiều sẽ gây mất nhiều thời gian và xuất hiện những giá trị None.

>**LƯU Ý RẰNG:**

Trong việc dự đoán giá nhà, việc **gộp hai (hay nhiều) nhóm nhà ở hai (hay nhiều) thành phố/tỉnh thành khác nhau** có thể không phải là lựa chọn tốt nhất. Dưới đây là một số lý do:

1. **Sự khác biệt về thị trường**: Mỗi thành phố có `những yếu tố kinh tế, xã hội, và bất động sản khác nhau ảnh hưởng đến giá nhà`. Ví dụ, giá nhà ở Hà Nội và TP.HCM có thể khác nhau rất nhiều do sự khác biệt về cung cầu, cơ sở hạ tầng, trình độ phát triển kinh tế và các yếu tố khác.

2. **Đặc trưng riêng biệt**: Các đặc trưng ảnh hưởng đến giá nhà (như tiện ích, vị trí, an ninh) có thể có mức độ ảnh hưởng khác nhau ở mỗi thành phố. Gộp dữ liệu từ hai (hay nhiều) thành phố có thể làm giảm tính chính xác của mô hình.

3. **Phân phối dữ liệu**: Việc gộp dữ liệu có thể tạo ra sự phân phối dữ liệu không đồng nhất, dẫn đến mô hình bị thiên lệch và không phản ánh đúng thực tế.

>**MỘT SỐ GIẢI PHÁP:**

1. **Huấn luyện mô hình riêng biệt**: Tạo và huấn luyện các mô hình dự đoán giá nhà riêng biệt cho từng thành phố. Sau đó, có thể so sánh kết quả hoặc hợp nhất kết quả từ các mô hình này nếu cần.

2. **Phân tích đa khu vực**: Nếu muốn sử dụng một mô hình duy nhất, cần thêm các biến phân loại đại diện cho từng thành phố và đảm bảo mô hình có khả năng xử lý sự khác biệt giữa các thành phố.

Trong việc giải quyết các bất động sản ở hai (hay nhiều) khu vực khác nhau, nếu số lượng bất động sản ở một khu vực khá ít so với tổng thể, ta có thể xem xét loại bỏ khu vực có ít dữ liệu hơn. Tuy nhiên, việc này phụ thuộc vào mục tiêu cụ thể của chúng ta và tác động của việc loại bỏ dữ liệu này. Dưới đây là một số điểm cần xem xét:  

1. **Khi Nên Loại Bỏ Dữ Liệu**:

    *  **Khác biệt lớn về đặc trưng**: Nếu hai khu vực có đặc trưng khác nhau rõ rệt, việc giữ lại dữ liệu của cả hai khu vực có thể làm phức tạp mô hình.
    *  **Số lượng dữ liệu quá ít**: Khu vực có quá ít dữ liệu có thể không đủ để tạo ra một mô hình học máy chính xác.

2. **Khi Không Nên Loại Bỏ Dữ Liệu**:

    *  **Thiếu đa dạng dữ liệu**: Loại bỏ dữ liệu của một khu vực có thể làm giảm đa dạng của dữ liệu, dẫn đến mô hình bị thiên lệch.
    *   **Mục tiêu nghiên cứu rộng**: Nếu bạn cần một mô hình có thể áp dụng cho nhiều khu vực, việc giữ lại tất cả dữ liệu có thể quan trọng.

>>2. **Mức giá không phù hợp với các đặc trưng vốn có của bất động sản** (vd: giá 36000 tỷ là quá cao so với một bất động sản có diện tích chưa đến 200 m^2, số tầng còn không vượt quá 5), có những mức giá quá thấp như 0.0036 tỷ (tức 3,6 triệu) nhưng xét về các đặc trưng vốn có như diện tích thì không phù hợp với mức giá bán như thế, bởi vì có một số bản tin là `CHO THUÊ` nên mức giá sẽ chỉ dao động từ `hàng triệu đến chục triệu` đây không phải là mức giá để bán một bất động sản ở khu vực phát triển như Thành Phố Hồ Chí Minh.

Tôi sẽ xét thêm điều kiện để loại bỏ những mức `GIÁ CHO THUÊ` này.

**Do đó chúng tôi sẽ kết luận các bản tin có mức giá không phù hợp là không đáng tin cậy và thêm điều kiện để loại bỏ**

````python
data_hcm['Price'].value_counts()
````

````python
# Đếm số lượng các giá trị nhỏ hơn 0.1 'tỷ' (tức 100 triệu) trong cột 'Price'
price_below_0_1 = data_hcm['Price'][data_hcm['Price'] < 0.1].count()
price_below_0_1
````

Như đã thấy thì `số lượng mức giá dưới 100 triệu là 68`, `tức có 68 bản tin là CHO THUÊ BẤT ĐỘNG SẢN`.



`Do không phù hợp với mục tiêu dự án là dự đoán giá nhà bán` nên tôi loại bỏ chúng khỏi tập dữ liệu của mình.

````python
# Hàm chia giá trị cho 1 tỷ nếu nằm trong khoảng từ 9 đến 12 chữ số (Tức ở mức Trăm Triệu - Trăm Tỷ)
def reprocess_price(price):
    try:
        price = float(price)
        # Chuyển đổi số thành chuỗi chỉ lấy phần nguyên
        integer_part = str(int(price))
        if 9 <= len(integer_part) <= 12:
            return price / 1000000000
        # Điều kiện để loại bỏ các mức giá không phù hợp, lớn hơn 3 chữ số
        elif len(integer_part) > 3:
            return None
        # Điều kiện để loại bỏ các mức giá CHO THUÊ (triệu/chục triệu)
        elif price < 0.1:
            return None
        else:
            return price
    except:
        return None
````

````python
data_hcm['Price'] = data_hcm['Price'].apply(reprocess_price)

# Loại bỏ các hàng có giá trị 'Price' là None
data_hcm = data_hcm.dropna(subset=['Price'])
````

###Target Encoding

Mã hóa biến mục tiêu (target encoding) là một kỹ thuật hữu ích trong học máy, đặc biệt khi làm việc với các biến mục tiêu phân loại (categorical targets).

Khi nào thì nên mã hóa biến mục tiêu:

1. **Biến mục tiêu là phân loại (Categorical Target):** Khi biến mục tiêu là phân loại, bạn cần chuyển đổi các nhãn (labels) thành các giá trị số để các mô hình học máy có thể xử lý được. Điều này thường được thực hiện với các biến mục tiêu phân loại, ví dụ như phân loại nhị phân (0 và 1) hoặc phân loại đa lớp.

2. **Tránh vấn đề phân loại không thứ tự (Unordered Classification):** Trong trường hợp biến mục tiêu có nhiều loại (multi-class classification) và các lớp không có thứ tự rõ ràng (e.g., 'red', 'green', 'blue'), việc sử dụng One-Hot Encoding có thể tạo ra nhiều đặc trưng và gây ra vấn đề với các mô hình có độ phức tạp cao. Mã hóa biến mục tiêu giúp đơn giản hóa vấn đề này.

3. **Cải thiện hiệu suất của mô hình:** Target encoding có thể cải thiện hiệu suất của các mô hình, đặc biệt là khi có các biến phân loại với nhiều giá trị duy nhất. Nó giúp giảm bớt số lượng các đặc trưng cần xử lý và có thể làm cho các mẫu trở nên dễ học hơn đối với các mô hình.

4. **Biến mục tiêu liên tục (Continuous Target):** Khi biến mục tiêu là liên tục (e.g., giá nhà), mã hóa bằng các phương pháp như Box-Cox transformation hoặc log transformation có thể giúp biến phân phối của biến mục tiêu trở nên gần với phân phối chuẩn, cải thiện hiệu suất của các mô hình hồi quy.

**Tuy nhiên, khi biến mục tiêu của tôi ở đây đang là một biến liên tục như Price, việc mã hóa biến mục tiêu thường không được áp dụng theo cách truyền thống (Label Encoding, One-Hot Encoding, Target Encoding,...).**

Một "outlier" là một điểm hoặc tập hợp các điểm khác biệt so với các điểm khác. Đôi khi chúng có thể rất cao hoặc rất thấp. Thường thì nên phát hiện và loại bỏ các outlier. Bởi vì outlier là một trong những lý do chính dẫn đến mô hình ít chính xác.   


Vì vậy, loại bỏ chúng `thường` là một ý tưởng tốt. Kỹ thuật phát hiện và loại bỏ outlier mà tôi sẽ thực hiện được gọi là `kỹ thuật điểm IQR`. Thường thì outlier có thể được nhìn thấy bằng cách sử dụng biểu đồ box plot. Dưới đây là biểu đồ box plot. Trong tất cả các biểu đồ này, bạn có thể thấy một số điểm ở bên ngoài hộp, chúng không gì khác ngoài outlier.

Tuy nhiên, **đối với một dữ liệu có số lượng Outlier quá lớn chẳng hạn như dữ liệu chúng tôi đang thực hiện thì việc loại bỏ các Outlier đồng thời sẽ làm mất đi lượng lớn thông tin** ảnh hưởng rất nhiều đến các giá trị dự đoán.

**Xử lý outlier không phải lúc nào cũng đòi hỏi phải loại bỏ chúng khỏi tập dữ liệu**. Có nhiều phương pháp khác để xử lý outlier một cách hiệu quả mà không làm mất đi thông tin có giá trị.

**MỘT SỐ PHƯƠNG PHÁP XỬ LÝ OUTLIER PHỔ BIẾN:**

1. **Biến đổi dữ liệu (Data Transformation):** Biến đổi dữ liệu có thể giúp làm giảm tác động của outlier. Một số biến đổi phổ biến bao gồm:

    >*  Log Transformation: Áp dụng logarit để làm giảm phạm vi của dữ liệu.
    >*  Square Root Transformation: Áp dụng căn bậc hai.  
    >* Box-Cox Transformation: Một biến đổi có thể tự động tìm kiếm biến đổi phù hợp.  


2. **Capping/Flooring (Winsorization):**  Capping hay flooring là phương pháp giới hạn giá trị outlier vào một ngưỡng cụ thể, bằng cách thay thế các giá trị outlier bằng giá trị ngưỡng.



3. **Sử dụng các thuật toán mạnh (Robust Algorithms):** Một số thuật toán học máy được thiết kế để ít bị ảnh hưởng bởi outlier, chẳng hạn như:

    >* Random Forest
    >* Gradient Boosting
    >* Support Vector Machines (SVM) với hàm mất mát phù hợp



4. **Kỹ thuật tính năng (Feature Engineering):** Thay vì loại bỏ outlier, bạn có thể tạo thêm các tính năng (features) mới giúp mô hình học máy học tốt hơn từ dữ liệu:

    >* Tạo cờ (flag) để đánh dấu các giá trị outlier.
    >* Tạo các tính năng tương tác (interaction features) hoặc đa thức (polynomial features).




5. **Imputation (Thay thế giá trị):** Thay thế giá trị outlier bằng một giá trị thống kê như trung vị (median) hoặc trung bình (mean) của dữ liệu.

>**Đối với tập dữ liệu có số lượng Outlier quá lớn như trên**, chúng tôi sẽ lựa chọn sử dụng **phương pháp Feature Engineering** để xử lý các Outlier.


>Feature Engineering có thể là một phương pháp mạnh mẽ để xử lý outlier mà không loại bỏ chúng. Một cách tiếp cận phổ biến là tạo `các biến cờ (flag variables) để đánh dấu các outlier`, hoặc biến đổi dữ liệu để giảm ảnh hưởng của chúng.

**Quy trình:**

* Tôi tạo các cột cờ (flag column) để đánh dấu các Outlier (việc xác định các Outlier dựa phương pháp tính `khoảng tứ phân vị IQR`)
* Các cột cờ mới đánh dấu các điểm Outlier của một đặc trưng (`1: là Outlier, 0: Không là Outlier`) xuất hiện trong Dataframe có tên nối đuôi là _outlier sau tên của Đặc trưng tương ứng (ví dụ: `Price_outlier`)

````python
import pandas as pd
import numpy as np

# Hàm xác định các phần tư và tính khoảng tứ phân vị (IQR)
def calculate_iqr_bounds(column):
    q1, q3 = np.percentile(column, [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return lower_bound, upper_bound

# Hàm tạo cột cờ (flag column) để đánh dấu outlier
def process_column(data, column_name):
    lower_bound, upper_bound = calculate_iqr_bounds(data[column_name])
    data[f'{column_name}_outlier'] = ((data[column_name] < lower_bound) | (data[column_name] > upper_bound)).astype(int)


# Danh sách các cột cần kiểm tra
columns_to_check = ['Price', 'Acreage', 'Bedrooms', 'Bathrooms', 'Floors']

# Thực hiện xử lý cho mỗi cột
for column in columns_to_check:
    process_column(data_hcm, column)


print(data_hcm.head())

# In thông tin về các cột cờ (flag columns)
for column in columns_to_check:
    print(f"{column} outliers: {data_hcm[f'{column}_outlier'].sum()}")

````

*Một số cách phổ biến để sử dụng các flag này trong quá trình phân tích và xây dựng mô hình:*

1. **Phân tích các outlier riêng biệt** : Bạn có thể tách riêng các hàng dữ liệu có chứa outlier để phân tích chi tiết hơn. Điều này giúp bạn hiểu rõ hơn về bản chất của các outlier và ảnh hưởng của chúng đến mô hình.

2. **Xử lý outlier bằng các phương pháp thay thế** : Thay vì loại bỏ các outlier, bạn có thể thay thế chúng bằng các giá trị khác như giá trị trung bình, trung vị hoặc sử dụng các phương pháp nội suy (interpolation).

3. **Sử dụng flag như một biến mới trong mô hình** : Thêm cột flag vào dữ liệu huấn luyện như một biến đầu vào mới. Điều này có thể giúp mô hình hiểu được sự hiện diện của các outlier và điều chỉnh dự đoán tương ứng.

4. **Tạo mô hình riêng cho outlier và non-outlier** : Bạn có thể tạo hai mô hình riêng biệt: một cho các hàng dữ liệu thông thường và một cho các hàng chứa outlier. Sau đó, kết hợp dự đoán từ hai mô hình này.

5. **Đánh giá mô hình dựa trên dữ liệu có và không có outlier** : Bạn có thể đánh giá hiệu quả của mô hình trên tập dữ liệu bao gồm cả outlier và không có outlier để xem sự khác biệt trong hiệu suất.


**Ở ĐÂY CHÚNG TÔI SẼ SỬ DỤNG CÁCH `Xử lý outlier bằng các phương pháp thay thế`**

>Đối với `Price` chúng tôi sẽ xử lý Outliers bằng cách fill median (lấp bằng trung vị) dựa theo `district` (Tức những điểm outlier sẽ được fill (điền) theo median (trung vị) của các điểm không phải outlier mà có cùng `district (quận/huyện)` )  


>Đối với `Acreage`, `Bedrooms`, `Bathrooms`, `Floors` chúng tôi sẽ chọn fill median theo các giá trị trung vị của mỗi cột.

**Xử lý các điểm outlier của `Price` bằng cách fill median của các điểm không phải outlier mà có cùng `district`**

````python
# Chức năng điền các giá trị ngoại lệ vào cột 'Giá' với trung vị quận
def fill_price_outliers(data):
    # Tính giá trung bình cho mỗi quận không bao gồm các giá trị ngoại lệ
    district_medians = data[data['Price_outlier'] == 0].groupby('district')['Price'].median()

    # Điền vào các giá trị ngoại lệ với giá trung bình của quận
    outlier_indices = data[data['Price_outlier'] == 1].index
    for idx in outlier_indices:
        district = data.at[idx, 'district']
        if district in district_medians:
            data.at[idx, 'Price'] = district_medians[district]

    return data
````

````python
data_hcm = fill_price_outliers(data_hcm)
````

**Xử lý outlier của các cột còn lại bằng cách điền theo trung vị của mỗi cột**

````python
# Hàm điền các ngoại lệ vào các cột được chỉ định bằng trung vị cột
def fill_outliers_with_column_median(data, columns):
    for column in columns:
        outlier_column = f'{column}_outlier'

        # Tính trung vị của cột hiện tại không bao gồm các giá trị ngoại lệ
        median_value = data.loc[data[outlier_column] == 0, column].median()

        # Điền vào các ngoại lệ với giá trị trung bình
        data.loc[data[outlier_column] == 1, column] = median_value

    return data

# Danh sách các cột cần xử lý
columns_fill_median = ['Acreage', 'Bedrooms', 'Bathrooms', 'Floors']
````

###Encode

>Mã hóa (encoding) các biến phân loại (categorical variables) là một bước quan trọng trong quá trình xử lý dữ liệu cho các mô hình học máy và phân tích dữ liệu.

**Một số lý do tại sao việc mã hóa là cần thiết:**

>1. **Đầu vào số hóa cho mô hình:**  Các mô hình học máy thường chỉ hoạt động tốt với dữ liệu số. Các thuật toán như hồi quy tuyến tính (Linear Regression), hồi quy logistic, SVM, và các mô hình dựa trên cây quyết định (Decision Tree) yêu cầu dữ liệu đầu vào là các giá trị số. Do đó, các biến phân loại cần được chuyển đổi thành dạng số để các mô hình này có thể xử lý được.

>2. **Tránh hiểu sai về quan hệ thứ tự:** Một số biến phân loại không có thứ tự tự nhiên (ví dụ: "red", "green", "blue" cho màu sắc). Nếu bạn sử dụng mã hóa số đơn giản (1, 2, 3), mô hình có thể hiểu sai rằng có sự thứ tự hay khoảng cách giữa các giá trị này. `One-Hot Encoding` giải quyết vấn đề này bằng cách tạo ra các cột nhị phân (binary columns) đại diện cho mỗi giá trị duy nhất trong biến phân loại, do đó loại bỏ mọi giả định về quan hệ thứ tự.

>3. **Tăng khả năng giải thích và diễn giải:** Sau khi mã hóa, mỗi biến được biểu diễn rõ ràng và dễ hiểu hơn. Điều này giúp cho việc diễn giải kết quả và các mối quan hệ trong dữ liệu trở nên dễ dàng hơn.

>4. **Tương thích với nhiều thuật toán:** Một số thuật toán học máy yêu cầu hoặc hoạt động tốt hơn với các biến phân loại đã được mã hóa. Ví dụ, thuật toán `k-NN (k-Nearest Neighbors)` hoạt động tốt hơn với dữ liệu đã được chuẩn hóa và mã hóa.

**Encoding Categorical Variables (Mã hóa các biến phân loại):**

>* **One-Hot Encoding**: Tạo ra các cột nhị phân (binary) riêng biệt cho mỗi giá trị trong biến phân loại.
>* **Label Encoding**: Chuyển đổi các giá trị phân loại thành các số nguyên.
>* **Binary Encoding**: Kết hợp giữa One-Hot Encoding và Label Encoding, phù hợp cho các biến phân loại có nhiều giá trị.

`Tôi sẽ KHÔNG THỰC HIỆN việc chuẩn hóa dữ liệu ngay ở phần này`

**Việc chuẩn hóa dữ liệu nên được thực hiện sau khi chia dữ liệu thành các tập huấn luyện (training) và kiểm tra (testing)**, với một số lý do sau:

>1. `Tránh rò rỉ thông tin (Data Leakage):` Chuẩn hóa toàn bộ dữ liệu trước khi chia có thể dẫn đến **rò rỉ thông tin từ tập kiểm tra vào tập huấn luyện**. Rò rỉ thông tin **làm mất tính khách quan của việc đánh giá mô hình**, do mô hình có thể đã học được thông tin từ tập kiểm tra.

## Feature Engineering

### Manual Information

>**Thông tin chung (Mutual Information (MI))** là một thước đo thống kê đo lường sự phụ thuộc giữa các biến. Nó giúp đánh giá mức độ thông tin mà một biến cung cấp về một biến khác. Trong ngữ cảnh tiền xử lý dữ liệu (preprocessing), MI có thể được sử dụng để chọn lọc các đặc trưng (feature selection) và đánh giá tầm quan trọng của chúng. Điều này đặc biệt hữu ích khi bạn có một tập dữ liệu lớn với nhiều đặc trưng và muốn chọn ra những đặc trưng có ích nhất cho mô hình của mình.

**Điểm Mutual Information (MI)** là một biện pháp thống kê dùng để đo lường độ tương quan giữa hai biến số.  

* Trong ngữ cảnh của học máy và phân tích dữ liệu, MI giúp xác định mức độ thông tin mà một biến số cung cấp về biến số khác.

* Khi sử dụng MI để chọn hoặc loại bỏ các đặc trưng trong một mô hình, **một số điểm MI bằng 0 cho biết rằng biến số đó không cung cấp bất kỳ thông tin hữu ích nào cho việc dự đoán biến mục tiêu**.  
Điều này có thể đủ cơ sở để xem xét loại bỏ đặc trưng đó.

>**Tuy nhiên, trước khi loại bỏ bất kỳ đặc trưng nào, nên thực hiện các bước kiểm tra và phân tích sâu hơn để xác nhận.**

````python
from sklearn.feature_selection import mutual_info_regression
````

````python
X = data_hcm.copy()
y = X.pop("Price")

# Mã hóa nhãn cho danh mục
for colname in X.select_dtypes("object"):
    X[colname], _ = X[colname].factorize()

# Tất cả các tính năng riêng biệt bây giờ phải có kiểu số nguyên (kiểm tra kỹ điều này trước khi sử dụng MI!)
discrete_features = X.dtypes == int
````

````python
# Hàm tính và trả về MI scores
def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores
````

````python
# Tính MI scores
mi_scores = make_mi_scores(X, y, discrete_features)
````

````python
mi_scores
````

````python
# Hàm Trực quan hóa MI scores
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")

# Vẽ biểu đồ trực quan
plt.figure(dpi=100, figsize=(10, 6))
plot_mi_scores(mi_scores)
plt.show()
````

### Correlation

#### Pearson

Đầu tiên và cơ bản nhất, tôi sẽ sử dụng hệ số tương quan của Pearson cho các đặc trưng số.

Hệ số tương quan Pearson đo lường mức độ liên quan tuyến tính giữa hai biến số. Giá trị của hệ số này nằm trong khoảng [-1, 1], với:

>* 1: Hai biến có mối tương quan thuận tuyệt đối.
>* -1: Hai biến có mối tương quan nghịch tuyệt đối.
>* 0: Hai biến không có mối tương quan tuyến tính.

Sử dụng thư viện có sẵn `pandas` và `seaborn` để tính toán và trực quan hóa.

````python
# Tính ma trận tương quan cho tất cả các cột
corr_matrix = data_hcm.corr()

# Thiết lập hình matplotlib
plt.figure(figsize=(15, 10))

# Sử dụng Seaborn để tạo bản đồ nhiệt
sns.heatmap(corr_matrix, annot= True, fmt = '.2f', cmap = 'Pastel2', linewidths = 2)

# Tiêu đề
plt.title('Correlation Heatmap')
````

>**Qua quan sát các hệ số tương quan, ta có thể thấy: `Longitude/Latitude`, `NearSchool` và `days_since_min` có quan hệ tuyến tính rất yếu / hầu như không có quan hệ tuyến tính với `Price`.**

````python
# Chuyển ma trận tương quan thành DataFrame
corr_pairs = corr_matrix.unstack()
# Lấy các cặp tương quan không phải với chính nó và sắp xếp theo độ lớn tương quan
sorted_corr_pairs = corr_pairs[(corr_pairs != 1)].sort_values(key=abs, ascending=False)
# Hiển thị top 10 cặp tương quan cao nhất
print(sorted_corr_pairs.head(10))
````

**Nhưng đôi khi việc sử dụng hệ số tương quan Pearson để đánh giá là không hiệu quả, ở đây tôi sẽ sử dụng một phương pháp đánh giá sự tương quan hiệu quả hơn (Lasso).**

#### Lasso

>`Lasso (Least Absolute Shrinkage and Selection Operator)`: là một phương pháp hồi quy có khả năng chọn lọc đặc trưng bằng cách đặt một số hệ số hồi quy bằng 0. Điều này rất hữu ích khi bạn muốn tìm ra những đặc trưng quan trọng nhất có ảnh hưởng đến biến mục tiêu.

>Lasso thực hiện điều này bằng cách `thêm một ràng buộc phạt vào hàm mất mát của hồi quy tuyến tính`, `cụ thể là tổng các giá trị tuyệt đối của các hệ số hồi quy`. Hình thức phạt này khuyến khích việc `nhiều hệ số hồi quy trở nên nhỏ hơn và có thể trở thành 0` (**GIẢM TRỌNG SỐ**), giúp giảm bớt các biến không quan trọng và đơn giản hóa mô hình.

**Dưới đây là cách sử dụng `Lasso` để tính toán các hệ số tương quan giữa các đặc trưng và biến mục tiêu `Price` trong `data_hcm`**

````python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
````

Chuẩn bị dữ liệu

````python
# Chọn các cột số cần chuẩn hóa
numerical_columns = data_hcm.drop(['Price'], axis=1).select_dtypes(include=['float64', 'int64']).columns

# Chuẩn hóa các cột số
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data_hcm[numerical_columns])


# Biến mục tiêu
y = data_hcm['Price']

# Tạo DataFrame cho các đặc trưng đã chuẩn hóa
X = pd.DataFrame(scaled_features, columns=numerical_columns)
````

Áp dụng Lasso

`Trước tiên phải tìm alpha tối ưu nhất`

````python
from sklearn.model_selection import GridSearchCV
# Phân chia dữ liệu thành tập huấn luyện và tập kiểm tra
# Tạo một đối tượng Lasso Regression
Lasso_regress = Lasso()

# Thiết lập các giá trị alpha muốn thử nghiệm
alphas = [0.0015,0.015, 0.025 ,0.15, 1, 10, 100]

# Tạo một grid search để tìm ra alpha tối ưu
param_grid = {'alpha': alphas}
grid_search = GridSearchCV(Lasso_regress, param_grid, cv = 5, scoring='neg_mean_squared_error')
grid_search.fit(X,y)
best_alpha = grid_search.best_params_['alpha']

# In ra giá trị alpha tối ưu và điểm số tương ứng
print("Best alpha for Lasso:", best_alpha)
print("Best MSE:", -grid_search.best_score_)
````

````python
# Khởi tạo mô hình Lasso với alpha tối ưu nhất
lasso = Lasso(alpha = best_alpha)

# Huấn luyện mô hình Lasso
lasso.fit(X, y)

# Lấy các hệ số của Lasso
lasso_coef = lasso.coef_

# Tạo DataFrame cho các hệ số Lasso
lasso_coef_df = pd.DataFrame({'Feature': numerical_columns, 'Lasso Coefficient': lasso_coef})
lasso_coef_df = lasso_coef_df.sort_values(by = 'Lasso Coefficient', ascending = False)
print(lasso_coef_df)
````

#### ĐA CỘNG TUYẾN (Multicollinearity)

**LƯU Ý:** Qua quan sát Heatmap có thể thấy hệ số tương quan của 2 đặc trưng `Bedrooms` và `Bathrooms` là `0.78` điều này có thể dẫn chúng ta đến vài điều cần lưu ý:

>* Một hệ số tương quan cao `(0.78)` giữa các cột `Bedrooms` và `Bathrooms` cho thấy rằng có một mối quan hệ mạnh mẽ giữa số phòng ngủ và số phòng tắm trong dữ liệu của bạn. Điều này có nghĩa là khi số lượng phòng ngủ tăng lên, số lượng phòng tắm cũng có xu hướng tăng lên, và ngược lại.

>* Trong bối cảnh mô hình hóa dự báo hoặc học máy, sự tương quan cao giữa các đặc trưng có thể dẫn đến vấn đề gọi là `đa cộng tuyến (multicollinearity)`. **Đa cộng tuyến có thể làm cho mô hình trở nên không ổn định và khó giải thích, do đó cần phải xử lý**.

**ĐA CỘNG TUYẾN**

**Đa cộng tuyến (multicollinearity)** là một vấn đề thường gặp trong hồi quy tuyến tính khi có mối tương quan mạnh giữa hai hoặc nhiều biến độc lập. Điều này có thể gây khó khăn trong việc xác định ảnh hưởng riêng biệt của từng biến độc lập lên biến phụ thuộc.

**Nguyên nhân:**

1. **Thiếu thông tin:** Nếu dữ liệu có ít quan sát hơn số lượng biến độc lập, khả năng xảy ra đa cộng tuyến cao.

2. **Biến tương tự:** Khi các biến độc lập mô tả các thông tin tương tự hoặc gần giống nhau, chúng sẽ có tương quan cao.

3. **Tạo biến mới từ biến cũ:** Sử dụng các phép biến đổi như bình phương hoặc log của các biến độc lập có thể dẫn đến đa cộng tuyến.

**Hậu quả:**

1. **Ước lượng không ổn định:** Hệ số hồi quy trở nên không ổn định và có thể thay đổi lớn khi thêm hoặc bớt biến.

2. **Giá trị p không chính xác:** Các kiểm định t có thể không đáng tin cậy, dẫn đến việc khó xác định biến nào thực sự có ý nghĩa.

3. **Sai lệch dự đoán:** Đa cộng tuyến làm tăng độ phức tạp của mô hình, gây khó khăn trong việc dự đoán chính xác.

**Kiểm tra đa cộng tuyến:**

1. **Hệ số tương quan:** Xem xét ma trận tương quan giữa các biến độc lập. Hệ số tương quan lớn hơn 0.8 hoặc nhỏ hơn -0.8 có thể chỉ ra đa cộng tuyến.

2. **VIF (Variance Inflation Factor):** Nếu VIF của một biến lớn hơn 10, điều này cho thấy biến đó có thể gây ra đa cộng tuyến.

3. **Eigenvalues và condition index:** Giá trị riêng nhỏ và chỉ số điều kiện lớn hơn 30 cho thấy có đa cộng tuyến.

**Giải pháp:**

1. **Loại bỏ biến:** Xóa bỏ một hoặc vài biến độc lập có tương quan cao.

2. **Kết hợp biến:** Kết hợp các biến có tương quan cao thành một biến tổng hợp.

3. **Biến đổi mô hình:** Sử dụng các kỹ thuật như hồi quy Ridge, Lasso để giảm thiểu ảnh hưởng của đa cộng tuyến.

4. **Thu thập thêm dữ liệu:** Nếu có thể, thu thập thêm dữ liệu để giảm thiểu tác động của đa cộng tuyến.

Cách kiểm tra đa cộng tuyến: Có một số phương pháp để kiểm tra đa cộng tuyến, trong đó phổ biến nhất là hệ số phóng đại phương sai (Variance Inflation Factor - VIF).

Hệ số phóng đại phương sai (VIF)
VIF đo lường mức độ tăng lên của phương sai của một hệ số hồi quy khi các biến độc lập có tương quan với nhau. VIF của một biến độc lập Xi được tính theo công thức:



image.png

trong đó R^2 là hệ số xác định khi Xi được hồi quy trên tất cả các biến độc lập khác trong mô hình.

Một quy tắc chung:

>* **VIF = 1**: Không có đa cộng tuyến giữa biến đó và các biến khác trong mô hình.

>* **1 < VIF < 5**: Mức độ đa cộng tuyến vừa phải. Có thể chấp nhận được trong nhiều trường hợp.

>* **VIF >= 5**: Đa cộng tuyến đáng kể. Cần xem xét kỹ lưỡng và có thể phải điều chỉnh mô hình.

>* **VIF >= 10**: Đa cộng tuyến rất cao. Đây là ngưỡng cảnh báo nghiêm trọng và thường đòi hỏi phải thực hiện các biện pháp để khắc phục vấn đề.

**CÁCH GIẢI QUYẾT:**

1. **Loại bỏ một trong hai đặc trưng:** có thể chọn loại bỏ một trong hai cột vì chúng mang thông tin tương tự nhau. Điều này giúp đơn giản hóa mô hình mà không mất quá nhiều thông tin.

2. **Tạo một đặc trưng mới kết hợp cả hai:**  có thể tạo một đặc trưng mới bằng cách kết hợp cả hai cột. Ví dụ, có thể tạo một cột mới gọi là Rooms tổng số phòng ngủ và phòng tắm.

3. **Sử dụng phương pháp phân tích thành phần chính (PCA):**  PCA là một kỹ thuật giảm chiều giúp chuyển đổi các đặc trưng có tương quan cao thành một tập hợp các đặc trưng không tương quan (thành phần chính). Điều này giữ lại phần lớn thông tin của dữ liệu ban đầu nhưng giảm thiểu vấn đề đa cộng tuyến.

Tôi sẽ chọn Kinh độ/Vĩ độ của trung tâm của `QUẬN 1` : `106.695833 / 10.776111`

**LÝ DO:**
>*  Quận 1 là nơi tập trung nhiều cơ quan chính quyền, các lãnh sự quán các nước và các tòa nhà cao tầng của thành phố (tòa nhà cao nhất Quận 1 và thứ nhì Thành phố Hồ Chí Minh là Bitexco Financial Tower). Quận 1 được xem là nơi sầm uất và có mức sống cao nhất của Thành phố về mọi phương diện, nổi tiếng với những địa điểm Phố đi bộ Nguyễn Huệ, Dinh Độc Lập, Nhà thờ chính tòa Đức Bà Sài Gòn, Thảo Cầm Viên Sài Gòn, Phố đi bộ Bùi Viện, Chợ Bến Thành,...
>*  Quận 1 là trung tâm hành chính của Thành phố Hồ Chí Minh, cũng là quận nổi tiếng và thu hút nhiều khách du lịch nhất của thành phố này. Khu vực trung tâm này tọa lạc rất nhiều khách sạn, homestay với đủ mức giá, phục vụ nhu cầu lưu trú của khách du lịch để tham quan các điểm du lịch trong khu vực hoặc di chuyển sang các quận lân cận.

Để thực hiện việc tính khoảng cách (đơn vị km) giữa hai tọa độ thì tôi sẽ sử dụng công thức `Haversine`. Công thức này giúp tính khoảng cách giữa hai điểm trên bề mặt của một quả cầu (trái đất).

````python
# Hàm tính khoảng cách theo công thức Haversine
def haversine(lon1, lat1, lon2, lat2):
    # Chuyển đổi độ sang radian
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    # Công thức Haversine
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))

    # Bán kính của Trái đất (trung bình) là 6371 km
    km = 6371 * c
    return km

# Tọa độ trung tâm Quận 1 (Longitude: 106.695833, Latitude: 10.776111)
center_longitude = 106.695833
center_latitude = 10.776111

# Tính khoảng cách và tạo đặc trưng mới 'Distance'
data_hcm['Distance_to_center'] = data_hcm.apply(lambda row: haversine(row['Longitude'], row['Latitude'], center_longitude, center_latitude), axis=1)
````

````python

# Đổi tên cột 'Amenities' thành 'Amenities_rating' với giá trị mặc định là giá trị 0
data_hcm = data_hcm.rename(columns={"Amenities": "Amenities_rating"})

# Tôi sẽ đổi tên Amenities thành Amenities_rating và gán các giá trị 0 (Thay cho việc loại bỏ Amenities và thêm đặc trưng mới Amenities _rating)
data_hcm['Amenities_rating'] = 0

# Gán trọng số cho các đặc trưng
weights = {
    'NearSchool': 1.2,  # Trọng số cho NearSchool
    'NearHospital': 1.8,  # Trọng số cho NearHospital
    'Security': 1.5  # Trọng số cho Security
}

# Tạo cột 'Amenities_rating' với trọng số
data_hcm['Amenities_rating'] += data_hcm['NearSchool'] * weights['NearSchool']
data_hcm['Amenities_rating'] += data_hcm['NearHospital'] * weights['NearHospital']
data_hcm['Amenities_rating'] += data_hcm['Security'] * weights['Security']
````

SAU KHI THỰC HIỆN KẾT HỢP TÔI LOẠI BỎ `Security, NearHospital, NearSchool`

````python
data_hcm = data_hcm.drop(['NearHospital', 'NearSchool', 'Security'], axis = 1)
data_hcm
````

### Represent The Correlation (Biểu diễn lại sự tương quan)

#### Pearson

Đầu tiên và cơ bản nhất, tôi sẽ sử dụng hệ số tương quan của Pearson cho các đặc trưng số.

Hệ số tương quan Pearson đo lường mức độ liên quan tuyến tính giữa hai biến số. Giá trị của hệ số này nằm trong khoảng [-1, 1], với:

>* 1: Hai biến có mối tương quan thuận tuyệt đối.
>* -1: Hai biến có mối tương quan nghịch tuyệt đối.
>* 0: Hai biến không có mối tương quan tuyến tính.

Sử dụng thư viện có sẵn `pandas` và `seaborn` để tính toán và trực quan hóa.

````python
# Tính ma trận tương quan cho tất cả các cột
corr_matrix = data_hcm.corr()

# Thiết lập hình matplotlib
plt.figure(figsize=(15, 10))

# Sử dụng Seaborn để tạo bản đồ nhiệt
sns.heatmap(corr_matrix, annot= True, fmt = '.2f', cmap = 'Pastel2', linewidths = 2)

# Tiêu đề
plt.title('Correlation Heatmap')
````

>**Qua quan sát các hệ số tương quan, có thể nhận thấy: `days_since_min` và `Distance_to_center` có quan hệ tuyến tính rất yếu với `Price`.**

````python
# Chuyển ma trận tương quan thành DataFrame
corr_pairs = corr_matrix.unstack()
# Lấy các cặp tương quan không phải với chính nó và sắp xếp theo độ lớn tương quan
sorted_corr_pairs = corr_pairs[(corr_pairs != 1)].sort_values(key=abs, ascending=False)
# Hiển thị top 10 cặp tương quan cao nhất
print(sorted_corr_pairs.head(10))
````

**Nhưng đôi khi việc sử dụng hệ số tương quan Pearson để đánh giá là không hiệu quả, ở đây tôi sẽ sử dụng một phương pháp đánh giá sự tương quan hiệu quả hơn.**

#### Lasso

`Lasso (Least Absolute Shrinkage and Selection Operator)`: là một phương pháp hồi quy có khả năng chọn lọc đặc trưng bằng cách đặt một số hệ số hồi quy bằng 0. Điều này rất hữu ích khi bạn muốn tìm ra những đặc trưng quan trọng nhất có ảnh hưởng đến biến mục tiêu.

**Dưới đây là cách sử dụng `Lasso` để tính toán các hệ số tương quan giữa các đặc trưng và biến mục tiêu `Price` trong `data_hcm`**

````python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
````

Chuẩn bị dữ liệu

````python
# Chọn các cột số cần chuẩn hóa
numerical_columns = data_hcm.drop(['Price'], axis=1).select_dtypes(include=['float64', 'int64']).columns

# Chuẩn hóa các cột số
scaler = StandardScaler()
scaled_features = scaler.fit_transform(data_hcm[numerical_columns])


# Biến mục tiêu
y = data_hcm['Price']

# Tạo DataFrame cho các đặc trưng đã chuẩn hóa
X = pd.DataFrame(scaled_features, columns=numerical_columns)
````

Áp dụng Lasso

`Trước tiên phải tìm alpha tối ưu nhất`

````python
from sklearn.model_selection import GridSearchCV
# Phân chia dữ liệu thành tập huấn luyện và tập kiểm tra
# Tạo một đối tượng Lasso Regression
Lasso_regress=Lasso()

# Thiết lập các giá trị alpha muốn thử nghiệm
alphas = [0.0015,0.015, 0.025,  0.15, 1, 10, 100]

# Tạo một grid search để tìm ra alpha tối ưu
param_grid = {'alpha': alphas}
grid_search = GridSearchCV(Lasso_regress, param_grid, cv = 5, scoring='neg_mean_squared_error')
grid_search.fit(X,y)
best_alpha = grid_search.best_params_['alpha']

# In ra giá trị alpha tối ưu và điểm số tương ứng
print("Best alpha for Lasso:", best_alpha)
print("Best MSE:", -grid_search.best_score_)
````

````python
# Khởi tạo mô hình Lasso với alpha tối ưu nhất
lasso = Lasso(alpha = best_alpha)

# Huấn luyện mô hình Lasso
lasso.fit(X, y)

# Lấy các hệ số của Lasso
lasso_coef = lasso.coef_

# Tạo DataFrame cho các hệ số Lasso
lasso_coef_df = pd.DataFrame({'Feature': numerical_columns, 'Lasso Coefficient': lasso_coef})
lasso_coef_df = lasso_coef_df.sort_values(by = 'Lasso Coefficient', ascending = False)
print(lasso_coef_df)
````

**Trực quan hóa các hệ số Lasso**

````python
plt.figure(figsize = (12, 8))
sns.barplot(x = 'Lasso Coefficient', y = 'Feature', data = lasso_coef_df, palette = 'viridis')
plt.title('Lasso Coefficients for Features')
plt.xlabel('Coefficient Value')
plt.ylabel('Feature')
plt.show()
````

#### SAU KHI KẾT HỢP `NearSchool, NearHospital, Security, Amenities` THÀNH `Amenities_rating` VÀ KẾT HỢP `Longitude, Latitude` THÀNH `Distance_to_center` THÌ:

> `Amenities_rating` **có tương quan DƯƠNG RẤT YẾU**, việc một đặc trưng có tương quan âm và một đặc trưng khác có tương quan dương kết hợp lại ra hệ số tương quan bằng GẦN 0 là hoàn toàn hợp lý và có thể xảy ra trong một số trường hợp.   

* Dù có hệ số tương quan rất yếu với biến mục tiêu nhưng theo kiến thức chuyên ngành tôi nhận thấy `Amenities_rating` là một đặc trưng quan trọng có tác động đến `Price` và dù tương quan rất yếu nhưng cũng có trọng số ảnh hưởng phần nào đến hiệu quả mô hình dự đoán, nên tôi vẫn sẽ giữ lại `Amenities_rating`

> `Distance_to_center` **có tương quan ÂM RẤT YẾU**, việc không có gì lạ vì ngay trước khi Kết Hợp thì `Longitude đã có tương quan dương rất yếu ` và `Latitude có tương quan là 0`, nhưng dù RẤT YẾU tôi vẫn sẽ quyết định giữ lại vì nó vẫn có ảnh hưởng đến kết quả dự đoán...(Dù rất ít)

VIỆC NAY ĐƯA TÔI ĐẾN MỘT QUYẾT ĐỊNH LÀ SẼ `KHÔNG LOẠI BỎ Amenities_rating` VÀ ` Distance_to_center`

### PCA(Principal Component Analysis)

`PCA(Principal Component Analysis)` (cũng là một trong những phương pháp Feature Engineering) đây là một kỹ thuật giảm chiều dùng để biến đổi các biến có tương quan với nhau thành một tập hợp các biến mới không có tương quan gọi là các thành phần chính (principal components). Các thành phần chính này giữ lại phần lớn sự biến thiên trong dữ liệu gốc. PCA thường được sử dụng để giảm chiều dữ liệu trong các bài toán học máy và trực quan hóa dữ liệu.

>Tuy nhiên ở phần `Standardization` chúng tôi đã quyết định chuẩn hóa dữ liệu sau khi đã chia thành các tập train và test.

>**Hơn nữa, thực hiện PCA (Phân tích Thành phần Chính) mà không chuẩn hóa dữ liệu trước có thể dẫn đến kết quả không chính xác vì các thành phần chính sẽ bị chi phối bởi các đặc trưng có đơn vị lớn hơn hoặc phạm vi giá trị lớn hơn.**

>Mặt khác, qua quá trình Preprocessing và Feature Engineering thì số lượng biến đặc trưng trong tập dữ liệu của chúng tôi đã giảm đi đáng kể, và cũng đã xử lý các biến tương quan cao với nhau.
Đôi khi việc thực hiện PCA là không cần thiết với một số trường hợp nhất định.

>Việc quyết định có nên sử dụng PCA cho tập dữ liệu của chúng tôi hay không sẽ phụ thuộc ở phần Methodology NẾU thực hiện PCA mà vẫn giữ được một kết quả dự đoán tốt và giảm đi đáng kể thời gian huấn luyện mô hình.
